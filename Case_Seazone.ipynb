```python
# Importando bibliotecas que serão utilizadas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leitura dos arquivos através da biblioteca pandas
df_ratings = pd.read_csv('desafio_ratings.csv',sep=";",encoding='latin-1')
df1_details = pd.read_csv('desafio_details.csv',sep=";",encoding='latin-1')
```

```python
# Retirando a coluna 'unnamed: 0'
df_ratings = df_ratings.drop(columns={'Unnamed: 0'})
df1_details = df1_details.drop(columns={'Unnamed: 0'})
```

### **3. Quais cidades têm mais avaliações?**

```python
# Mesclando as tabelas ratings e details para conseguir fazer cruzamento de dados entre elas
df_merge = pd.merge(df_ratings, df1_details, on='hotel_id', how='inner')
```

```python
df_merge.groupby('city_name')['number_of_ratings'].sum().sort_values(ascending=False)
```

## **4. Quais cidades têm a maior média de avaliações? E a menor média? a. Existe alguma explicação para isso? Conjecture.**

```python
# Mapeando as notas para a escala de 1 a 10
df_merge['Total'] = df_merge['Total'].map({10: 10, 9: 9, 8: 8, 7: 7, 6: 6, 5: 5, 4: 4, 3: 3, 2: 2, 1: 1})

# Calculando a média das avaliações por cidade
med_avaliacoes = df_merge.groupby('city_name')['Total'].mean().reset_index()

# Filtrando e ordenando as cidades com as maiores médias (entre 8 e 10)
maiores_medias = med_avaliacoes[(med_avaliacoes['Total'] >= 8) & (med_avaliacoes['Total'] <= 10)].sort_values(by='Total', ascending=False)
maiores_medias
```

```python
# Filtrando e ordenando as cidades com as menores médias (entre 1 e 3)
menores_medias = med_avaliacoes[(med_avaliacoes['Total'] >= 1) & (med_avaliacoes['Total'] <= 3)].sort_values(by='Total', ascending=False)
menores_medias
```

```python
# Conta o número de avaliações por anúncio
contagem_avaliacoes_por_anuncio = df_merge['number_of_ratings'].value_counts()

# Filtra anúncios que têm pelo menos 3 avaliações
anuncios_com_3_ou_mais_avaliacoes = contagem_avaliacoes_por_anuncio[contagem_avaliacoes_por_anuncio >= 3].index

# Filtra o DataFrame para incluir apenas anúncios com pelo menos 3 avaliações
df_filtrado = df_merge[df_merge['number_of_ratings'].isin(anuncios_com_3_ou_mais_avaliacoes)]

# Calcula a média de avaliações considerando apenas anúncios com pelo menos 3 avaliações
media_geral = df_filtrado['Total'].mean()
media_geral
```

A nota é basicamente fruto da experiencia do hóspede com ao imóvel e tendo em vista que a satisfação média dos hóspedes é alta. O que tenho como explicação é o fato dos casos de menor avaliação se da por a cidade ter poucos um apenas um imóvel para locação e esse imóvel ter sido alugado uma ou pouquíssimas vezes e a experiencia pontual desse(s) hóspede(s) foi negativa e com uma baixa amostragem acaba afundando a nota de determinado anúncio.

### **5. Existem correlações entre as características de um anúncio e a sua localização? a. Quais? Explique**

```python
# Cria um mapa de calor para visualizar as correlações nos dados utilizando o seaborn
matriz_correlacao = df_merge.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(matriz_correlacao, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Matriz de Correlação')
plt.show()
```

Não, como mostra a matriz acima, a localização tem correlação neutra e/ou ligeramente negativa com as características de cada imóvel/anúncio (qtd_facilities e room_surface_in_m2), isto é, a localizalção não é afetada pelas características de cada imóvel/anúncio e vice-versa.

### **6. Existem relações entre a nota do anúncio e os recursos disponíveis no imóvel? a. Quais? Explique.**

```python
# Análise descritiva
desc = df_merge.describe()
# Removendo columas que não acrescentaram na análise
desc = desc.drop(columns=['hotel_id', 'room_id'])
desc
```

```python
# Calcula o comprimento das listas de recursos
df_merge['qtd_facilities'] = df_merge['room_facilities'].apply(len)

# Cria um gráfico de dispersão para visualizar a relação entre nota e quantidade de recursos
plt.figure(figsize=(12, 8))
sns.scatterplot(x='qtd_facilities', y='Total', data=df_merge)
plt.title('Relação entre Nota do Anúncio e Quantidade de Recursos')
plt.xlabel('Quantidade de Recursos')
plt.ylabel('Nota do Anúncio')
plt.show()
```

```python
correlacao = df_merge['Total'].corr(df_merge['qtd_facilities'])
print(f"Correlação entre a quantidade de recursos e a nota do anúncio é de: {correlacao}")
```

Não, a correção entre a nota e a recursos disponíveis é extremamente baixa, ou seja, não ha influencia entre elas.
